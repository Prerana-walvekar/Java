1 --> fraudulent transaction

The dataset is highly imbalanced because fraudulent transactions are very rare compared to legitimate ones


 If the model sees too many normal transactions, it might always predict "normal" and miss fraud.


```python
# separating the data for analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions -->
 492

randomly selects 492 legitimate transactions from the dataset.

Since there are 492 fraud cases, we take 492 legitimate transactions to balance the dataset.

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)#if we give axis=0 added one after another(row)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

The sampled transactions show clear differences in feature values between fraud and legitimate transactions. This helps the model learn fraud patterns better.

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

X now contains only the features (V1, V2, ..., Amount, etc.).

Y contains 0 for legit transactions and 1 for fraud transactions.

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

Training on 80% of data helps the model learn patterns.
Testing on 20% of data helps check if the model works on unseen data.

Training Phase : 
The model learns patterns from X_train (input) and Y_train (output).


Testing Phase : 
The model makes predictions on X_test.
Predictions are compared with Y_test to check accuracy.

stratify=Y :- same proportion
random_state=2 :- split remains the same every time the code runs 

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

 Logistic Regression is simple, efficient, and well-suited for binary classification (fraud or not fraud). It requires less computational power and performs well when combined with data balancing techniques.

```python
model = LogisticRegression()
```

We use StandardScaler to make sure all features in the dataset have the same scale.

If we don’t scale them, bigger numbers might dominate the model.

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

X_train → Contains the input features (scaled values of transactions).
Y_train → Contains the labels (0 for legit transactions, 1 for fraud).
model.fit() → The model learns patterns from X_train to predict Y_train

Model Evaluation

Accuracy Score

Calculates accuracy as:

Accuracy= 

Correct Predictions/Total Predictions
​
 
Gives an idea of how well the model has learned from the training data.

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

conclusion:

The Logistic Regression model achieved 94.28% training accuracy and 91.37% test accuracy, showing good performance.

The small accuracy drop indicates low overfitting
